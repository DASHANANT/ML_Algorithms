# Input Features

```python
features=df2.iloc[:,:-1]
target=df2.iloc[:,-1]
```

## 1. Standardize the Dataset


```python
features= (features-features.mean())/features.std()
features
```

# 2. Calculate the covariance matrix for the whole dataset

# Since we have standardized the dataset,
# so the mean for each feature is 0 
# and the standard deviation is 1.

```python
covmat=features.cov()
covmat
```

# 3. Calculate eigenvalues and eigen vectors.

```python
C=covmat.to_numpy()
C
```

```python
from numpy import linalg as LA
eigenvalue, eigenvector = LA.eig(C)
```

```python
eigenvalue
```

```python
s=np.sum(eigenvalue)
s
```

```python
eigenvector 
```

# 4. Sort eigenvalues and their corresponding eigenvectors.

already in sorted order

```python

for i in range(len(eigenvalue)):
    print("k",i,"variance",np.sum(eigenvalue[:i+1])/s)
```

# 5. Pick k eigenvalues and form a matrix of eigenvectors
## If we choose the top 18 eigenvectors, the matrix will look like this:
## which give variance upto 80%

```python
eigenvalue[:10]
```

```python
kcomp=eigenvector[:10].T
kcomp.shape
```

# 6. Transform the original matrix.
Feature matrix * top k eigenvectors = Transformed Data

```python
Transformed_Data=np.dot(features,kcomp)
```

```python
Transformed_Data.shape
```

```python
target=np.array(target)
y=pd.DataFrame(target)
y
```

```python
x = pd.DataFrame(Transformed_Data)
x
```

```python
x['10'] =target
```

```python
data=x.copy()
```

# Model -Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data
```

```python
def prepare_XY(df):        
    X = []
    Y = []
    df = np.array(df)
    m,n = df.shape      
    X = np.array(df[:,:n-1]) # X array except last column
    Y = np.array(df[:,n-1]) # Y array with last column
    return(X,Y)
```

```python
X, y = prepare_XY(data)
z=[]
k=[]
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    clf = LogisticRegression()
    clf.fit(X_train, y_train)
    clf.predict(X_test)
    clf.predict_proba(X_test)
    z.append(clf.score( X_test, y_test))
    k.append(clf.coef_)

idx=np.argmax(z)
print('score is ',z[idx]*100,'%')
print('')
print('weights are')
print(k[idx])
```
